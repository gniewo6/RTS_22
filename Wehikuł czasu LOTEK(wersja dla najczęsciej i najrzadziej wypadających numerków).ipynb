```python
import csv
from heapq import nlargest, nsmallest
import pandas as pd

def csv_to_M(f):    
    csv_f = csv.reader(f)
    M = []  #definiujemy pustą liste
    for i in csv_f: # for w którym i do element list, czyli lista z 6 elementami
        row = 49*[0] #generuje/resetuje liste zawierającą 49 zer
    
        for j in i: #każde "j" równa się trafionemu numerkowi
      #W pustym "row" nastepuje zmina tylko w indeksie opowiadającym numerkowi. 
      #Np. pierwszy numer "i" to 8, więc row[7]=1
          row[int(j)-1]=1 # na pocztku dałem int(j) by mieć lepszy wgląd
        M = M + [row] # dodaje kolejny element do listy list, zawsze zawiera 6 "1" i 43 "0"
    return M
    
def slownik(): #Generuje słownik który będzie wykorzystany by przypisywać wartości numerką
    my_dict = {}
    for i in range(49):
        my_dict[i+1] = 0
    return my_dict



def frequency_of_number(M):
    cn_dict = slownik()
    for m in M: #pętla idzie po macierzy i daje nam pojedyńczą liste 'm'
        petla = 1# licznik pętli(dla poniższej pętli) zaczynający od 1, reprezentuje numerek 
        for i in m:
            cn_dict[petla] += i #wartość "i" ma wartość 0 lub 1
            petla += 1 #"petla" zwiększa by dodać(lub nie zmienić) wartość dla danego numerka
#pod koniec 'petla' wraca do wartości 1
    return cn_dict

def largest_number (M): #Zwraca macierz rozmiarowo taką samą, ale z najczęściej wypadającymi numerkami
    Max = []
    for i in range(len(M)):
        x = frequency_of_number(M[:i]) 
        s_nc_w = nlargest(6, x, key = x.get) 
        Max.append(s_nc_w)
    return Max

def smallest_number (M): #jak wyżej tylko dla małych numerków
    Max = []
    for i in range(len(M)):
        x = frequency_of_number(M[:i]) 
        s_nc_w = nsmallest(6, x, key = x.get) 
        Max.append(s_nc_w)
    return Max

def comper_M(M1,M2):# z dwóch macierzy robi nam listę zgodnych numerków, każdy element to zgodność w jednym losowaniu
    lM1 = len(M1)
    if lM1 == len(M2):# za mało zgłębiłem metode "try, excepts" by jej użyć
        Mcomper = []
        for i in range(lM1):
            z = []
            for j in range(49): #dałem 49 ponieważ długoś M[0] = 49
                z.append(M1[i][j] and M2[i][j])
            Mcomper.append(z)
    
        return Mcomper

    else:
        return 'Error one muszą być równo liczne'
```

```python
l = open('lotek1973.csv')
M = csv_to_M(l)
#troche to zajmuje
df =  pd.DataFrame(smallest_number(M))#tworzy plik .csv, który będzie porównywany z wynikami z prawdziwego świata
df.to_csv('najrzadziej występujące.csv', sep=',', header=None, index=None)

df =  pd.DataFrame(largest_number(M))
df.to_csv('najczesciej występujące.csv', sep=',', header=None, index=None)
```

```python
j = open('najrzadziej występujące.csv')
k = open('najczesciej występujące.csv')
M_nw6 = csv_to_M(k)
M_nr6 = csv_to_M(j)

#W dalszych wersjach poniższy kod dostanie własną funkcje
zuzia = comper_M(M,M_nw6)
dziubi = []
print('wehikuł czasu dla niajcześciej wypadających')
for i in zuzia:
    dziubi.append(sum(i))
for i in range(7):
    print(i,dziubi.count(i))

print()    
print('wehikuł czasu dla niajrzadziej wypadających')
new_M = comper_M(M,M_nr6)
ziubi = []
for i in new_M:
    ziubi.append(sum(i))
for i in range(7):
    print(i,ziubi.count(i))

```
