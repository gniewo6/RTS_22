```python
import csv

f = open('lotek.csv')
csv_f = csv.reader(f)
M = []  #definiujemy pustą liste
print(csv_f)
for i in csv_f: # for w którym i do element list, czyli lista z 6 elementami
    row = 49*[0] #generuje/resetuje liste zawierającą 49 zer
    
    for j in i: #każde "j" równa się trafionemu numerkowi
      #W pustym "row" nastepuje zmina tylko w indeksie opowiadającym numerkowi. 
      #Np. pierwszy numer "i" to 8, więc row[7]=1
      row[int(j)-1]=1 # na pocztku dałem int(j) by mieć lepszy wgląd
    M = M + [row] # dodaje kolejny element do listy list, zawsze zawiera 6 "1" i 43 "1"
```

```python

lw = 2000 #losowania wstecz które uwzględnimy
cn =  49*[0]  # Częstość wypadania numerków w totka
for j in range(49):
    kolumna = 0
    for i in range(lw,0, -1):
        kolumna = M[i][j] + kolumna
        #print(kolumna)
    cn[j] = kolumna/lw
    
print(cn)
```

```python
cnx = list(cn) #kopiowanie listy
sum_x = 0
for i in cnx:
    sum_x = round((max(cnx)*100/6),3) + sum_x
    print((cnx.index(max(cnx))+1), round((max(cnx)*100/6),3), "%")
    cnx[cnx.index(max(cnx))] = 0
print(sum_x)
```

```python
for i in range(lw): #for sprawdzajacy czy zawsze jest 6 jedynek w wierszu
    if(M[i].count(1)!=6):
        print("M[", i,"]")
```

```python
print(M)
with open('output.csv', 'w', newline="") as f: # tworzy plik output.csv 
    writer = csv.writer(f)
    writer.writerows(M) # lista list "M" jest wklejana do outputu
```
